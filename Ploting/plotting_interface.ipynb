```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
#read csvfile output of project
df1 = pd.read_csv('progress-BNN.csv')
```

```python
#read csvfile output of project
df2 = pd.read_csv('progress-GNN.csv')
```

```python
df1
```

```python
df_loss1 = df1.drop('accuracy', 1)
df_loss1 = df_loss1.dropna()

df_acc1 = df1.drop(['loss', 'batch'], 1)
df_acc1 = df_acc1.dropna()

```

```python
df_loss2 = df2.drop('accuracy', 1)
df_loss2 = df_loss2.dropna()

df_acc2 = df2.drop(['loss', 'batch'], 1)
df_acc2 = df_acc2.dropna()
```

```python
fig, axs = plt.subplots(2, 1)
fig.set_size_inches(10,7)


#axs[0].set_title('Loss')
axs[0].plot(df_loss1['epoch'], df_loss1['loss'],label='BN Centralized',color="#06C2AC")
axs[0].plot(df_loss2['epoch'],df_loss2["loss"],color="#8C000F",label='GN Centralized')
axs[0].legend(prop={'size': 18})
axs[0].set_ylabel('Loss', fontsize=25)

axs[0].tick_params(axis='x',labelsize=15)
axs[0].tick_params(axis='y',labelsize=15)
#axs[1].set_title('Accuracy')



axs[0].grid(True);
axs[1].set_ylim((0,100))
axs[1].plot(df_acc1['epoch'], df_acc1['accuracy'] * 100, color="#06C2AC",label='BN  Centralized')
axs[1].plot(df_acc2['epoch'], df_acc2['accuracy'] * 100,color="#8C000F",label='GN  Centralized')
axs[1].legend(prop={'size': 18},loc='lower right')
axs[1].set_ylabel('Accuracy', fontsize=25)
axs[1].set_xlabel('Epochs', fontsize=25)
axs[1].tick_params(axis='x',labelsize=15)
axs[1].tick_params(axis='y',labelsize=15)
fig.tight_layout()
axs[1].grid(True);
```
